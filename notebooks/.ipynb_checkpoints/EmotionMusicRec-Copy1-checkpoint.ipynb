# Emotion-Based Music Recommender System

In this project, we develop a system that leverages Natural Language Processing (NLP) to detect emotions from textual input and recommends music based on these emotions using the Spotify API. We use a pre-trained DistilBERT model for emotion classification and integrate with the Spotify API for music recommendations.


## Setup and Imports

```python
# Importing necessary libraries for the project
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments
from datasets import load_dataset
import numpy as np
import torch
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

## Loading the data

```python
# Loading the emotion dataset with a specified split for training and validation
dataset = load_dataset("emotion", split={'train': 'train[:50%]', 'validation': 'validation[:10%]'})
```

## Data Exploration

Before diving into model training, it's essential to explore our dataset. This will give us insights into the distribution of emotions, the length of text inputs, and other characteristics that could influence our approach to training the model.


```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Converting the dataset to a pandas DataFrame
df_train = pd.DataFrame(dataset['train'])

# Extracting emotion names from the dataset
emotion_counts = dataset['train'].features['label'].names

# Plotting the distribution of emotions
plt.figure(figsize=(10, 6))
sns.countplot(x='label', data=df_train)
plt.title('Emotion Distribution in the Training Set')
plt.xlabel('Emotion')
plt.xticks(ticks=range(len(emotion_counts)), labels=emotion_counts, rotation=45)
plt.ylabel('Number of Instances')
plt.show()
```

```python
# Calculating the length of each text entry
text_lengths = [len(text.split()) for text in dataset['train']['text']]

# Plotting the distribution of text lengths
plt.figure(figsize=(10, 6))
sns.histplot(text_lengths, bins=30)
plt.title('Distribution of Text Lengths')
plt.xlabel('Text Length (Number of Words)')
plt.ylabel('Frequency')
plt.show()
```

## Tokenization

```python
# Preprocessing the text data using the tokenizer from the 'distilbert-base-uncased' model
tokenizer = AutoTokenizer.from_pretrained('distilbert-base-uncased')

def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True, max_length=128)

tokenized_datasets = dataset.map(tokenize_function, batched=True)
```

## Model Intialization

```python
# Initializing the model for sequence classification with 6 emotion labels
model = AutoModelForSequenceClassification.from_pretrained('distilbert-base-uncased', num_labels=6)
```

## Training Setup

```python
# Setting up training arguments to define the training process
training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=3,
    per_device_train_batch_size=8,
    evaluation_strategy="epoch",
    save_strategy="epoch",
    load_best_model_at_end=True,
)

# Initializing the trainer with the model, training arguments, and datasets
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets['train'],
    eval_dataset=tokenized_datasets['validation'],
)
```

## Model Training

```python
# Training the model on the emotion dataset
trainer.train()
```

## Saving the Model and Tokenizer

```python
# Saving the trained model and tokenizer for future use
model_path = "./emotion_detection_model"
model.save_pretrained(model_path)
tokenizer.save_pretrained(model_path)
```

## Unique Labels

```python
# Displaying the unique emotion labels from the dataset
unique_labels = dataset['train'].features['label'].names
print("Unique labels:", unique_labels)
```

## Emotion Prediction on Custom Text

```python
# Predicting the emotion of a custom text input
text = "I am feeling happy today"
inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=128)

with torch.no_grad():
    logits = model(**inputs).logits

probabilities = torch.softmax(logits, dim=1)
predicted_class_id = probabilities.argmax().item()
class_names = ['sadness', 'joy', 'love', 'anger', 'fear', 'surprise']
predicted_emotion = class_names[predicted_class_id]

print(f"Predicted Emotion: {predicted_emotion}")
```

## Evaluation on Test Data

```python
# Evaluating the model's performance on the test dataset
dataset_testing = load_dataset("emotion", split='test')
test_texts = dataset_testing['text']
test_labels = dataset_testing['label']

test_encodings = tokenizer(test_texts, padding=True, truncation=True, max_length=128, return_tensors="pt")

with torch.no_grad():
    outputs = model(**test_encodings)
predictions = torch.argmax(outputs.logits, dim=1)

accuracy = (predictions == torch.tensor(test_labels)).float().mean()
print(f"Test Accuracy: {accuracy.item()}")
```

## Results Overview

```python
# Comparing actual and predicted emotion labels in a DataFrame
df_results = pd.DataFrame({'Actual Label': [unique_labels[label] for label in test_labels[:10]],
                           'Predicted Label': [unique_labels[pred] for pred in predictions[:10].numpy()]})
print(df_results)
```

## Confusion Matrix Visualization

```python
# Visualizing the model's performance across all categories with a confusion matrix
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

cm = confusion_matrix(test_labels, predictions.numpy(), normalize='true')
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=unique_labels)
disp.plot(cmap=plt.cm.Blues)
plt.title('Normalized Confusion Matrix')
plt.show()
```
